# Data prep

Imputation: median (to be optimized)

```python
def prep_data(df,inplace=False):
    if inplace==False:
        df=df.copy()
    
    df.Age.fillna(df.Age.median(),inplace=True)
    df.fillna(df.mode().iloc[0],inplace=True)
    
    df['Title'] = df.Name.apply(lambda x: x.split()[1])
    df.Title = df.Title.apply(lambda x: x if x in ['Mr.','Miss.','Mrs.','Master.','Dr.'] else 'etc')
    
    drop_col = ['Name','Ticket','Cabin']
    df.drop(drop_col,axis=1,inplace=True)
    
    df = pd.get_dummies(df,columns=['Sex','Embarked','Title'])
#     df = pd.get_dummies(df,columns=['Sex','Embarked'])
    
    drop_col = ['Sex_female','Embarked_C']
    df.drop(drop_col,axis=1,inplace=True)
    
    return df
```

# Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def model_train(df_train):
    
    x_train = df_train.iloc[:,1:]
    y_train = df_train.iloc[:,0]
    
    param_dict = {'max_depth':(3,4,5,6,7),'max_features':(0.3,0.5,0.8),
              'class_weight':(None,'balanced')}

    model = RandomForestClassifier(n_estimators=200,min_samples_split=2)
    model_gs = GridSearchCV(model,param_grid=param_dict, n_jobs=-1)

    model_gs.fit(x_train,y_train)
    
    print('Best_score:',model_gs.best_score_)
    print('Best_param:',model_gs.best_params_)
    return model_gs
```

```python
# Data prep
df_train_prep = prep_data(df_train)
df_test_prep = prep_data(df_test)
```

```python
# Train model
model_gs = model_train(df_train_prep)
```

```python
pred = model_gs.predict(df_test_prep)
df_test['Survived']=pred
# df_test.Survived.to_csv('result/result03.csv',header=True)
```
